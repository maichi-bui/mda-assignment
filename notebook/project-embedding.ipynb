```python
import re
import json
import nltk
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

# Download necessary NLP resources
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('punkt_tab')
```

```python
from sentence_transformers import SentenceTransformer

model = SentenceTransformer("sentence-transformers/all-MiniLM-L6-v2")
```

```python

# 1. Normalize text: Lowercase, remove special characters (except necessary punctuation)
def normalize_text(text):
    text = text.lower()
    text = re.sub(r"[^a-zA-Z0-9\s.,]", "", text)  # Keep only alphanumeric, spaces, and periods
    text = re.sub(r"\s+", " ", text).strip()  # Remove extra spaces
    return text

# 2. Tokenization
def tokenize_text(text):
    return word_tokenize(text)

# 3. Remove stopwords (Dutch and German stopwords for mixed-language text)
def remove_stopwords(tokens):
    stop_words = set(stopwords.words("dutch")) | set(stopwords.words("german"))  # Combine Dutch & German stopwords
    return [word for word in tokens if word not in stop_words and word.isalnum()]

# 4. Lemmatization (reduces words to base forms)
def lemmatize_tokens(tokens):
    lemmatizer = WordNetLemmatizer()
    return [lemmatizer.lemmatize(word) for word in tokens]

# 5. Sentence Tokenization for structured input
def structure_sentences(text):
    sentences = sent_tokenize(text)
    return sentences
```

```python
# get sample of data to test
sample_data = projects.sample(1000)[['id','title','objective']].to_dict('records')
```

```python
def generate_embeddings(prj):

  structured_sentences = structure_sentences(normalize_text(prj['objective']))
  structured_sentences.append(normalize_text(prj['title']))

  embeddings = model.encode(structured_sentences)
  mean_embedding = np.mean(embeddings, axis=0)

  return mean_embedding
```

```python
embeddings_sample = []
for prj in tqdm(sample_data):
  mean_embedding = generate_embeddings(prj)
  embeddings_sample.append(mean_embedding)
```

```python
len(embeddings_sample)
```

```python
def find_top_similar_embeddings(embedding, all_embeddings, top_k=5):
  """
  Calculates cosine similarity between a given embedding and a list of embeddings,
  returning the indices of the top_k most similar embeddings.

  Args:
    embedding: The query embedding (NumPy array).
    all_embeddings: A list of embeddings (NumPy arrays).
    top_k: The number of top similar embeddings to return.

  Returns:
    A list of indices of the top_k most similar embeddings.
  """
  similarities = [np.dot(embedding, other_embedding) / (np.linalg.norm(embedding) * np.linalg.norm(other_embedding))
                  for other_embedding in all_embeddings]
  top_indices = np.argsort(similarities)[::-1][:top_k+1]
  return top_indices.tolist()[1:]
```

```python
# Example usage:
# Assuming you have a list of embeddings called 'embeddings'
# and you want to find the top 5 similar embeddings for the first embedding:
query_embedding = embeddings_sample[100]
top_similar_indices = find_top_similar_embeddings(query_embedding, embeddings_sample, top_k=5)
print("Top 5 similar embeddings indices:", top_similar_indices)
pd.DataFrame(sample_data).iloc[top_similar_indices]
```
