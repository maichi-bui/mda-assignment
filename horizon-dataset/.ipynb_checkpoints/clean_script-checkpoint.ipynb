# clean data

## clean project.xlsx

```python
project['frameworkProgramme'].nunique()
# drop this column contain only 1 value "HORIZON"
project.drop(columns=['frameworkProgramme'], inplace=True)
```

```python
project[~project.nature.isnull()]
# drop this column contain all "na"
project.drop(columns=['nature'], inplace=True)
```

```python
# these two columns are identical, drop 1 of them
project[project['masterCall']!=project['subCall']]
project.drop(columns=['subCall'], inplace=True)
```

```python
def convert_num(x):
    try:
        num = int(x)
        
        return num
    except ValueError:
        try:
            num = float(x.replace(',','.'))
            return num
        except:
            print(f"This is not a number. {x}")
            return x
project['ecMaxContribution'] = project['ecMaxContribution'].apply(convert_num)
project['totalCost'] = project['totalCost'].apply(convert_num)    
```

```python
project.startDate=pd.to_datetime(project['startDate'])
project.endDate=pd.to_datetime(project['endDate'])
```

```python
project.set_index('id',inplace=True)
```

```python
profile = ProfileReport(project, title="Projects file EDA")
# profile.to_notebook_iframe()
profile.to_file("prj_data.html")
```

```python
project.to_csv('cleaned-data/projects.csv')
```
